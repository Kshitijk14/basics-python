* Supervised learning deals with or learns with “labeled” data. This implies that some data is already tagged with the correct answer.
* 

## Categories
* 

1. Classifcation => when the output variable is a category, such as “Red” or “blue” , “disease” or “no disease”
2. Regression => when the output variable is a real value, such as “dollars” or “weight”

## Types

1. Regression (linear, non-linear, polynomial, multiple linear)
2. Logistic Regression
3. Classification
4. Naive Bayesian Model
5. K-NN (k nearest neighbors)
6. Decision Trees
7. Random Forest
8. Support Vector Machine
